```
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os

plt.rcParams['figure.figsize'] = (16, 8)

#Si la imagen es RGB, la convierte a B y N
A = imread('test.bmp')
X = np.mean(A, -1)

#Se crean 3 variales que contienen U(columnas) S(Diagonales) y VT(Filas

U, S, VT = np.linalg.svd(X, full_matrices=False)
S = np.diag(S)

j = 0

#Se hace uso del SVD en algebra lineal, para crear una matriz de tama√±o mxn con base en los valores U, S, VT
for i in (None, None, 200):
     Xapprox = U[:, :i] @ S[0:i, :i] @ VT[:i, :]
     plt.figure(j+1)
     j += 1
     img = plt.imshow(Xapprox)
     plt.axis('off')
     img.set_cmap('gray')
     plt.title('i =' + str(i))
     plt.savefig("CompressedImg.jpg")
     plt.show()

#Compress Ratio

OGsize = os.stat("test.bmp").st_size
CSize = os.stat("CompressedImg.jpg").st_size
CompRatio = (CSize/OGsize)*100
print("El peso de la imagen original es:", OGsize)
print("El peso de la imagen original es:", CSize)
print("la imagen comprimida es un:", round(CompRatio), "% de la original")
```
